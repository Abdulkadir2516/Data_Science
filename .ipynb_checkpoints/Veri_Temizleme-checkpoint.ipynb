```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# basit uyarıları gösterme
import warnings
warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("./Data_sets/athlete_events.csv")
veri.head()
```

```python
# veri hakkında bana bilgi veriyor
veri.info()
```

```python
# veri sütunlarını görüntüleme
veri.columns
```

```python
# sütun isimlerini değiştirme
veri.rename(columns={
    "ID":"id",
    "Name":"isim",
    "Sex":"cinsiyet",
    "Age":"yas",
    'Height': "boy", 
    'Weight': "kilo", 
    'Team': "takim", 
    'NOC': "uok", 
    'Games': "oyunlar",
    'Year': "yil", 
    'Season': "sezon", 
    'City': "sehir", 
    'Sport': "spor", 
    'Event': "etkinlik", 
    'Medal': "madalya"}, inplace = True)

veri.head(2)
```

```python
# işimize yaramayan sütunları çıkarıyoruz:
veri = veri.drop(["id","oyunlar"], axis =1)# axis =1 sütun manasında
veri.head(2)
```

```python
# benzersiz olan etkinlikleri yeni değişkene alıyoruz
unique_etkinlik = pd.unique(veri.etkinlik)

print("Benzersiz olan etkinliklerin sayısı => " , unique_etkinlik.size)
unique_etkinlik[:10]
```

```python
# her bir etkinliği iteratif olarak dolaşıp 
# etkinlik özelinde boy ve kilo ortalamalarını hesaplayacağız
# etkinlik özelinde kayıp verileri ortlamaya göre eşitle

veri_pass = veri.copy()

boy_kilo_liste = ["boy","kilo"]

for i in unique_etkinlik:
    # etkinlik filtre
    e_filtre = veri_pass.etkinlik == i
    
    filter_data = veri_pass[e_filtre]
    
    for x in boy_kilo_liste:
        
        ort = np.round(np.mean(filter_data[x]))
        
        # her bir etkinlikte bazı boş değerler varsa yani hepsi boş değilse dolu olanlara göre boşları doldur        
        if ~np.isnan(ort): 
            filter_data[x] = filter_data[x].fillna(ort)
        else: # etkinlik üzerinde hepsi boş ise tüm tablodaki dolu verilere göre doldur
            tum_ort = np.round(np.mean(veri_pass[x]))
            filter_data[x] = filter_data[x].fillna(tum_ort)

    # işlemleri yaptıktan sonra geçici olan verimize aktaralım
    veri_pass[e_filtre] = filter_data
    
# kayırp verilerden arındırılmış olan geçici verimi gerçek veriye eşitleme

veri = veri_pass
```

```python
veri_pass.info()
```

```python
# yas değerlerini nan olanları bulma doldurma

yas_ort = np.round(np.mean(veri_pass["yas"]))
veri_pass["yas"] = veri_pass["yas"].fillna(yas_ort)
veri_pass.info()
```

```python
madalya_degiskeni = veri["madalya"]
pd.isnull(madalya_degiskeni).sum()
```

```python
madalya_filtresi = pd.notnull(madalya_degiskeni)

veri_pass = veri_pass[madalya_filtresi]
veri_pass.info()
```

```python
veri_pass
```

```python
# Bütün Temizlenmiş Verilerimiz
veri = veri_pass
veri
```

```python
veri.to_csv("./Data_sets/athlete_events_clear.csv", index = False)
```
