```python
def plotHistogram(degisken):
    """
        Girdi: Degişken/Sütun ismi
        Çıktı: İlgili değişkenin histogramı
    """
    plt.figure()
    plt.hist(veri[degisken], bins =100 , color="green" )
    plt.xlabel(degisken)
    plt.ylabel("Frekans")
    plt.title("Veri Sıklığı - {}".format(degisken))
    plt.show()
    
```

```python
numerik_degiskenler = ["yas","boy","kilo","yil"]

for i in numerik_degiskenler:
    plotHistogram(i)
```

```python
def plotbar(degisken, n=5):
    """
        Girdi: Değişken / Sütun ismi
                n = Gösterilecek eşsiz değer sayısı
        Çıktı: Çubuk grafiği
    """
    veri_ = veri[degisken]
    veri_sayma = veri_.value_counts()
    veri_sayma = veri_sayma[:n]
    
    plt.figure()
    plt.bar(veri_sayma.index, veri_sayma, color="red")
    plt.xticks(veri_sayma.index, veri_sayma.index.values, rotation=90)
    plt.ylabel("Frekans")
    plt.title("Veri Sıklığı - {}".format(degisken))
    plt.show()
    
    print("{}: \n{}".format(degisken, veri_sayma))
    
```

```python
kategorik_degiskenler = ['isim', 'cinsiyet', 'takim', 'uok', 'sezon', 'sehir', 'spor', 'etkinlik', 'madalya']

for i in kategorik_degiskenler:
    
    plotbar(i,20)
    
```

```python
from collections import Counter

def anomaliTespiti(df, ozellik):
    
    outlier_indices=[]
    
    for c in ozellik:
        # 1. çeyrek
        q1 = np.percentile(df[c],25)
        
        #3. çeyrek
        q3 = np.percentile(df[c],75)
        
        # IQR = Inter Quartile Range
        iqr = q3-q1
        
        # aykırı değer için ek adım miktarı
        outlier_step = 1.5 * iqr
        
        # aykırı değerleri ve bulunduğu indeksi tespit etme
        outlier_list_col = df[(df[c] < q1 - outlier_step) | (df[c] > q3 + outlier_step)].index
        
        # tespit edilen indeksleri depolama
        outlier_indices.extend(outlier_list_col)
        
    #eşsiz aykırı değerleri bulma
    outlier_indices = Counter(outlier_indices)
    
    # eğer bir örnek n kadar sütunda farklı ise bu değeri aykırı kabul edelim
    
    multiple_outliers = list(i for i,n in outlier_indices.items() if n>1)
    
    return multiple_outliers
        
```

```python
veri_anomali = veri.loc[anomaliTespiti(veri,["yas","kilo","boy"])]
veri_anomali.spor.value_counts()
```

```python
# veri içerisinde bulunan yıl değerlerinin date time veri tipine dönüştürelim

tarih_saat_nesnesi = pd.to_datetime(veri_zaman["yil"], format="%Y")

print(type(tarih_saat_nesnesi))
tarih_saat_nesnesi.head()
```

```python
veri_zaman["tarih_saat"] = tarih_saat_nesnesi
veri_zaman.head(3)
```

```python
# ver aman değişkeninin ana indeksini date time olan tarih saat değerine eşitleyelim
veri_zaman = veri_zaman.set_index("tarih_saat")
veri_zaman.drop(["yil"], axis=1, inplace= True)
veri_zaman.head()
```

```python
periyodik_veri = veri_zaman.resample("2A").mean()
periyodik_veri.head()
```

```python
# kayıp verileri çıkaralım
periyodik_veri.dropna(axis=0, inplace=True)
periyodik_veri.head()
```

```python
plt.figure()
periyodik_veri.plot()
plt.xlabel("Yıl")
plt.title("Yıllara göre Ortalama Yaş, Boy ve Ağırlık")
plt.grid(True)
plt.legend()
plt.show()
```
